## Evaulate 




```python
from ultralytics import YOLO

model = YOLO("runs/detect/train3/weights/best.pt") 


metrics = model.val(
    data="basketball-1/data.yaml",
    split="test",
    conf=0.3,  
    iou=0.5
)
```

```python
basketball_idx = list(model.names.values()).index("basketball")

precision = metrics.box.p[basketball_idx]
recall = metrics.box.r[basketball_idx]
map50 = metrics.box.ap50[basketball_idx]
map95 = metrics.box.ap[basketball_idx]

f1 = 2 * (precision * recall) / (precision + recall + 1e-16)

print("\n=== Basketball Only Metrics ===")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"mAP@0.5: {map50:.4f}")
print(f"mAP@0.5:0.95: {map95:.4f}")
```
